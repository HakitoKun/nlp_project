```python
plt.plot(train_steps, train_loss)
```

```python
# Train Loss vs EvalLoss
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot()
ax.plot(train_steps, train_loss)
ax.plot(*zip(*eval_loss))
ax.xaxis.set_major_locator(plt.MaxNLocator(7))
ax.legend(['Train Loss', 'eval_loss'])
ax.set_xlabel("Steps")
ax.set_ylabel("Value Loss")
ax.set_title("Train Loss VS eval_loss")
```
